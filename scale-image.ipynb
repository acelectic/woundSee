```python
import glob, os
import cv2
import numpy as np
```

```python
old_width = 640
new_width = 1280
src_dir = 'set1'
dest_dir = 'scale-width-'+str(new_width)
dest_scale_down_dir = 'scale-down-width-'+str(new_width)

os.makedirs(dest_dir, exist_ok=True)
os.makedirs(dest_scale_down_dir, exist_ok=True)
```

```python
for img_path in glob.glob(src_dir + '/*/*.jpg'):
    img = cv2.imread(img_path)

    img_dir_name = img_path.split(os.sep)[-2]
    os.makedirs(os.path.join(dest_dir, img_dir_name), exist_ok=True)

    scale = new_width / img.shape[1]

    img = cv2.resize(img, None, fx=scale, fy=scale)

    new_img_path = os.path.join(dest_dir, img_dir_name,
                                img_path.split(os.sep)[-1])
    cv2.imwrite(new_img_path, img)
```

```python
scale_up_dir = glob.glob(dest_dir + '/*/*.jpg') + glob.glob(dest_dir + '/*/*.png') + glob.glob(dest_dir + os.path.join('\\unknow','*','*.jpg'))
# scale_up_dir = glob.glob(dest_dir + '/*/*')

for img_path in scale_up_dir:
#     print(img_path)
    img = cv2.imread(img_path)
    if(np.count_nonzero(img) > 0):
        img_dir_name = img_path.split(os.sep)[:-2]
        img_dir_name = os.sep.join(img_dir_name).replace(dest_dir, dest_scale_down_dir)
        print(img_dir_name)
        os.makedirs(os.path.join(dest_scale_down_dir, img_dir_name), exist_ok=True)

        scale = old_width / img.shape[1]

        img = cv2.resize(img, None, fx=scale, fy=scale)

        new_img_path = os.path.join(dest_scale_down_dir, img_dir_name,
                                    img_path.split(os.sep)[-1])
        cv2.imwrite(new_img_path, img)
```
